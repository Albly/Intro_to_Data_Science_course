```python
import copy
df_file_mod = copy.deepcopy(df_file)
```

```python
# Q15.1 Create a `sqft_tot_area` column (sum of all columns with `sqft_` prefix) using any method above
df_file_mod['sqft_tot_area'] = (df_file_mod.filter(regex='^sqft_',axis=1).apply(np.sum , axis=1))
df_file_mod
```

```python
# Q15.2 Create a new column `sqm_tot_area` using `sqft_tot_area` and the fact that 1 foot = 0.3048 meters
df_file_mod['sqm_tot_area'] = df_file_mod['sqft_tot_area'].apply(lambda x: x/(0.3048**2))
df_file_mod

#https://stackoverflow.com/questions/33768122/python-pandas-dataframe-how-to-multiply-entire-column-with-a-scalar
```

```python
  # Q15.3 Create a new column `sqm_aver_floor_area` by dividing total area (in meters) by number of floors
df_file_mod['sqm_aver_floor_area'] = df_file_mod['sqm_tot_area'].div(df_file_mod['floors'])
df_file_mod
```

```python
# Q15.4 Create a new column `price_cat` by splitting a `price` into 5 ([1..5]) distinct intervals: 0 < x <=20%,
# 20% < x <= 40%, ... 80% < x <= 100% percentiles. You could use `.quantile()` to compute percentiles.
df_file_mod['price_cat'] = pd.qcut(df_file_mod['price'],[0, 0.2, 0.4, 0.6, 0.8, 1],labels=False)
```

```python
# Q15.5 Create a new bool column `high_class` it is True if the house has grade >= 9 and condition >= 4
df_file_mod['high_class'] = (df_file['grade'] >= 9 ) & (df_file['condition'] >= 4)
df_file_mod
```

```python
# Q22 Split your data into train and test parts.
# How many records (rows) do you have in train and test tables? (list of int)?
# Use sklearn.model_selection.train_test_split with test_size=0.33 and random_state=7

#here we choose features
#df_file_mod = df_file_mod.query("price < 4500000")

num_features = ['bedrooms','bathrooms', 'sqft_living', 'sqft_lot', 'floors' , 'sqft_above' , 'sqft_basement']
cat_features = ['waterfront', 'view', 'condition', 'grade', 'house_age_10','year','month' ]

X = df_file_mod[[*num_features, *cat_features]]
Y = df_file_mod['price']
```

```python
# Preprocessing of categorical features
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
encoded_features = pd.DataFrame(ohe.fit_transform(X[['waterfront', 'view', 'condition', 'grade', 'house_age_10','year','month' ]]))
encoded_features

waterfront_cols = [f'waterfront_{col}' for col in ohe.categories_[0]]
view_cols = [f'view_{col}' for col in ohe.categories_[1]]
condition_cols = [f'condition_{col}' for col in ohe.categories_[2]]
grade_cols = [f'grade_{col}' for col in ohe.categories_[3]]
decade_cols = [f'decade_{col}' for col in ohe.categories_[4]]
year_cols = [f'year_{col}' for col in ohe.categories_[5]]
month_cols = [f'month_{col}' for col in ohe.categories_[6]]

dummy_col_names = [*waterfront_cols,*view_cols,*condition_cols,*grade_cols,*decade_cols,*year_cols,*month_cols]
encoded_features.columns = dummy_col_names

encoded_features
```

```python
# changing categorical features to encoded data
X.drop(['waterfront', 'view', 'condition', 'grade', 'house_age_10','year','month' ], axis= 1, inplace=True)
X = pd.concat([X, encoded_features], axis=1)
X
```

```python
#Splitting dato into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = .33, random_state=7)
print(X_train.shape)
print(X_test.shape)
print([X_train.shape[0], X_test.shape[0]]) # rows that we have in train and test tables (list of int)
```
